## cosine graph 예측

```python
import torch
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import math

# 데이터 생성
x = torch.linspace(-math.pi, math.pi, 1000)
y = torch.cos(x)

# 무작위 매개변수 초기화
a = torch.randn(()) * 0.01
b = torch.randn(()) * 0.01
c = torch.randn(()) * 0.01
d = torch.randn(()) * 0.01
e = torch.randn(()) * 0.01
f = torch.randn(()) * 0.01

# 무작위 그래프 생성
y_random = a * x**3 + b * x**2 + c * x + d + e * x**4 + f * x**5

# 그래프 그리기
plt.subplot(2, 1, 1)
plt.title('cosine graph')
plt.plot(x, y)
plt.subplot(2, 1, 2)
plt.title('random cosine graph')
plt.plot(x, y_random)
plt.show()
```

```python
# 데이터 로더 설정
dataset = TensorDataset(x, y)
batch_size = 50
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# 학습률 설정
learning_rate = 1e-7
```

```python
# 학습 루프
for epoch in tqdm(range(50000)):  # tqdm 사용
    for batch_x, batch_y in dataloader:
        y_pred = a * batch_x**3 + b * batch_x**2 + c * batch_x + d + e * batch_x**4 + f * batch_x**5
        loss = (y_pred - batch_y).pow(2).sum().item()  # 평균 사용

        grad_y_pred = 2.0 * (y_pred - batch_y)
        grad_a = (grad_y_pred * batch_x ** 3).sum()
        grad_b = (grad_y_pred * batch_x ** 2).sum()
        grad_c = (grad_y_pred * batch_x).sum()
        grad_d = grad_y_pred.sum()
        grad_e = (grad_y_pred * batch_x ** 4).sum()
        grad_f = (grad_y_pred * batch_x ** 5).sum()
        
        a -= learning_rate * grad_a
        b -= learning_rate * grad_b
        c -= learning_rate * grad_c
        d -= learning_rate * grad_d
        e -= learning_rate * grad_e
        f -= learning_rate * grad_f

    if epoch % 100 == 0:
        print(f'epoch:{epoch+1} loss:{loss}')
```

```python
y_pred = a * x**3 + b * x**2 + c * x + d + e * x**4 + f * x**5
```

```python
plt.subplot(3, 1, 1)
plt.title("y true")
plt.plot(x, y)
plt.subplot(3, 1, 2)
plt.title("y pred")
plt.plot(x, y_pred)
plt.subplot(3, 1, 3)
plt.plot(y_random)
plt.title("y random")
plt.show()
```
